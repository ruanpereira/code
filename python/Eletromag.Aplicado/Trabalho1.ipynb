```python
import sympy as sp
import matplotlib as mpl
from sympy.vector import CoordSys3D, Del, gradient, curl, divergence
```

```python
# Questão 01
# item A

# Trazendo o frame C (sympy precisa disso pra funcionar), basicamente criando as coordenadas cartesianas para
# realizar a equação
R = sp.vector.CoordSys3D('R')

# Definindo V, o uso de R.x e R.z é para delimitar quais são as coordenadas de campo da equação, 
# para que o Sympy faça as derivações da operação gradiente. 
V_a = 3 * R.x**2 + R.x * R.z

print('Item a')

# Operação gradiente aplicada no campo
grad_a = sp.vector.gradient(V_a)
print("Gradiente")
print(grad_a)

# Operação Laplaciano, também conhecido como sendo o divergente do gradiente, sendo assim aplico a divergência
# no resultado do gradiente
lap_a = sp.vector.divergence(grad_a)
print('Laplaciano')
print(lap_a)

# Operação rotacional
rot_a = sp.vector.curl(grad_a)
print('Rotacional')
print(rot_a)

# item B

# Trazendo o frame C (sympy precisa disso pra funcionar), basicamente criando as coordenadas cilíndricas para
# realizar a equação
C = sp.vector.CoordSys3D('C', transformation='cylindrical', variable_names=('ro', 'phi', 'z'))

# Definindo V, o uso de R.x e R.z é para delimitar quais são as coordenadas de campo da equação, 
# para que o Sympy faça as derivações da operação gradiente. 
V_b = C.ro *  C.z * (sp.cos(C.phi))

print('Item b')

# Operação gradiente aplicada no campo
grad_b = sp.vector.gradient(V_b)
print('Gradiente')
print(grad_b)

# Operação Laplaciano, também conhecido como sendo o divergente do gradiente, sendo assim aplico a divergência
# no resultado do gradiente
lap_b = sp.vector.divergence(grad_b)
print('Laplaciano')
print(lap_b)

# Operação rotacional
rot_b = sp.vector.curl(grad_b)
print('Rotacional')
print(rot_b)

# item C

# Trazendo o frame E (sympy precisa disso pra funcionar), basicamente criando as coordenadas esféricas para
# realizar a equação
E = sp.vector.CoordSys3D('E', transformation='spherical', variable_names=('r', 'theta', 'phi' )  )

# Definindo V, o uso de R.x e R.z é para delimitar quais são as coordenadas de campo da equação, 
# para que o Sympy faça as derivações da operação gradiente. 
V_c = 4*E.r**2*(sp.cos(E.theta))*(sp.sin(E.phi))

print('Item C')

# Operação gradiente aplicada no campo
grad_c = sp.vector.gradient(V_c)
print('Gradiente')
print(grad_c)

# Operação Laplaciano, também conhecido como sendo o divergente do gradiente, sendo assim aplico a divergência
# no resultado do gradiente
lap_c = sp.vector.divergence(grad_c)
print('Laplaciano')
print(lap_c)

# Operação rotacional
rot = sp.vector.curl(grad_c)
print('Rotacional')
print(rot)
```

```python
#Questão 05
# item a
import numpy as np

# Vetores encontrados no gráfico do item A
x = np.arange(0,11,1) # cm
V = np.array([0,2,3,3,2,0,-2,-3,-3,-2,0])

E = (-1)*np.diff(V)
```

```python
# Organizando o vetor para plotagem
E = np.hstack((E,E[1]))
```

```python
# Criando o gráfico
from matplotlib import pyplot as plt
plt.plot(x,E*1e-3)
plt.grid()
plt.ylabel("E [J]")
plt.xlabel("x [cm]")

plt.savefig("q5a.pdf")
```
